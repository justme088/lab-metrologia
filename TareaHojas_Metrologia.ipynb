```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el CSV
df = pd.read_csv('/content/drive/MyDrive/GoogleColab/Metrologia/hojas_datos.csv')
```

```python
# Calcular estadísticas
largo_promedio = df['largo_cm'].mean()
largo_std = df['largo_cm'].std()
largo_rango = df['largo_cm'].max() - df['largo_cm'].min()

ancho_promedio = df['ancho_cm'].mean()
ancho_std = df['ancho_cm'].std()
ancho_rango = df['ancho_cm'].max() - df['ancho_cm'].min()

print("Largo de hojas:")
print(f"  Promedio: {largo_promedio:.2f} cm")
print(f"  Desviación estándar: {largo_std:.2f} cm")
print(f"  Rango: {largo_rango:.2f} cm")

print("\nAncho de hojas:")
print(f"  Promedio: {ancho_promedio:.2f} cm")
print(f"  Desviación estándar: {ancho_std:.2f} cm")
print(f"  Rango: {ancho_rango:.2f} cm")
```

```python
# Histograma de largo
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(df['largo_cm'], bins=10, color='green', edgecolor='black')
plt.title('Histograma del Largo de las Hojas')
plt.xlabel('Largo (cm)')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
# Histograma de ancho
plt.subplot(1, 2, 2)
plt.hist(df['ancho_cm'], bins=10, color='blue', edgecolor='black')
plt.title('Histograma del Ancho de las Hojas')
plt.xlabel('Ancho (cm)')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
